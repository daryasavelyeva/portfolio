### Проверка расчета эффективности обогащения

Проверим, что эффективность обогащения рассчитана правильно:

- Вычислим её на обучающей выборке для признака `rougher.output.recovery`
- Найдем *MAE* между нашими расчётами и значением признака

```python
C = data_train['rougher.output.concentrate_au'] # доля золота в концентрате после флотации/очистки
F = data_train['rougher.input.feed_au'] # доля золота в сырье/концентрате до флотации/очистки
T = data_train['rougher.output.tail_au'] # доля золота в отвальных хвостах после флотации/очистки
```

```python
# формула расчета эффективности обогащения
recovery = (C * (F - T)) / (F * (C - T)) * 100
```

```python
print(f"Среднеее значение эффективности обогащения = {recovery.mean()}")
print(f"MAE = {mean_absolute_error(data_train['rougher.output.recovery'], recovery)}")
```

Значение средней абсолютной ошибки стремится к нулю, значит, расчет произведен верно.

### Суммарная концентрация всех веществ на разных стадиях: в сырье, в черновом и финальном концентратах

Посчитаем суммарную концентрацию всех веществ на разных стадиях:

```python
# суммарная концентрация металлов в исходном сырье
feed_concentrate = (data_full['rougher.input.feed_au']
                    + data_full['rougher.input.feed_ag']  
                    + data_full['rougher.input.feed_pb'] 
                    + data_full['rougher.input.feed_sol'])

# суммарная концентрация металлов после флотации
rougher_concentrate = (data_full['rougher.output.concentrate_au'] 
                       + data_full['rougher.output.concentrate_ag'] 
                       + data_full['rougher.output.concentrate_pb'] 
                       + data_full['rougher.output.concentrate_sol'])

# суммарная концентрация металлов после первичной очистки
primary_cleaner_concentrate = (data_full['primary_cleaner.output.concentrate_au'] 
                               + data_full['primary_cleaner.output.concentrate_ag'] 
                               + data_full['primary_cleaner.output.concentrate_pb'] 
                               + data_full['primary_cleaner.output.concentrate_sol'])

# финальная суммарная концентрация металлов
final_concentrate = (data_full['final.output.concentrate_au'] 
                     + data_full['final.output.concentrate_ag'] 
                     + data_full['final.output.concentrate_pb'] 
                     + data_full['final.output.concentrate_sol'])
```

 - Мы видим, что суммарная концентрация веществ увеличивается с каждым этапом, на этапе флотации и финальном этапе концентрация похожа, однако, распределение в финальном концентрате более собранное

## Работа с моделями

### Разбивка данных на выборки

Добавим в обучающую и тестовую выборки целевые признаки из исходной выборки:

```python
data_train = pd.merge(data_train, 
                     data_full[['rougher.output.recovery', 'final.output.recovery']],
                     left_index=True,
                     right_index=True)

data_test = pd.merge(data_test, 
                     data_full[['rougher.output.recovery', 'final.output.recovery']],
                     left_index=True,
                     right_index=True)

data_train.info()
data_test.info()
```

Разделим обучающую выборку на признаки и целевой признак:

```python
# сохраним признаки
features_train = data_train.drop(['rougher.output.recovery', 'final.output.recovery'], axis=1)

# сохраним целевой признак
target_train = data_train[['rougher.output.recovery', 'final.output.recovery']]

display(features_train.shape)
display(target_train.shape)
```

Разделим тестовую выборку на признаки и целевой признак:

```python
# сохраним признаки
features_test = data_test.drop(['rougher.output.recovery', 'final.output.recovery'], axis=1)

# сохраним целевой признак
target_test = data_test[['rougher.output.recovery', 'final.output.recovery']]

display(features_test.shape)
display(target_test.shape)
```

### Функция для вычисления итоговой sMAPE

Создадим функцию для расчета метрики качества *sMAPE* (англ. *Symmetric Mean Absolute Percentage Error*, «симметричное среднее абсолютное процентное отклонение»), которая принимает на вход целевой признак и предсказания:

```python
def smape(target, predictions):
    return 1 / len(target) * np.sum(2 * np.abs(target - predictions) 
                                    / (np.abs(target) + np.abs(predictions)) * 100)
```

Также создадим функция для итоговой метрики *sMAPE*, которая складывается из двух величин:
- эффективность обогащения чернового концентрата `rougher.output.recovery`
- эффективность обогащения финального концентрата `final.output.recovery`

На вход функция также принимает целевой признак и предсказания:

```python
def final_smape(target, predictions):
    smape_rougher = smape(target['rougher.output.recovery'], predictions[:, 0])
    smape_final = smape(target['final.output.recovery'], predictions[:, 1])
    
    final_smape = (smape_rougher * 0.25) + (smape_final * 0.75)
    
    return final_smape
```

Создадим метрику оценки качества модели `smape_scorer`:

```python
smape_scorer = make_scorer(final_smape, greater_is_better=False)
```

### Обучение моделей и оценка их качества

Обучим и оценим качество трех моделей: *линейная регрессия*, *дерево решений* и *случайный лес*

**Линейная регрессия**

Обучим модель и оценим ее качество с помощью кросс-валидации:

```python
lr_model = LinearRegression()

score = cross_val_score(lr_model, features_train, target_train, cv=5, scoring=smape_scorer).mean()

print('Среднее значение sMAPE для модели LinearRegression = ', score)
```

**Дерево решений**

Обучим модель, используя автоматический подбор лучших гиперпараметров с помощью GridSearchCV, и оценим результат ее работы:

```python
# настроим псевдослучайность выбора
STATE = np.random.RandomState(12345)
```

```python
dtr_model = DecisionTreeRegressor(random_state=STATE)

params_dtr_model = {'max_depth': range(1, 11)}

grid_dtr_model = GridSearchCV(dtr_model, params_dtr_model, cv=5, scoring=smape_scorer)

grid_dtr_model.fit(features_train, target_train)

print('Лучшее значение sMAPE для модели DecisionTreeRegressor = ', grid_dtr_model.best_score_, 
      'при параметрах', grid_dtr_model.best_params_ )
```

**Случайный лес**

Аналогично обучим модель, используя автоматический подбор лучших гиперпараметров с помощью GridSearchCV, и оценим результат ее работы:

```python
rfr_model = RandomForestRegressor(random_state=STATE, n_jobs=-1)

params_rfr_model = {'n_estimators': range(2, 33, 5),
                    'max_depth': range(1, 10, 2)}

grid_rfr_model = GridSearchCV(rfr_model, params_rfr_model, cv=5, scoring=smape_scorer)

grid_rfr_model.fit(features_train, target_train)

print('Лучшее значение sMAPE для модели RandomForestRegressor = ', grid_rfr_model.best_score_, 
      'при параметрах', grid_rfr_model.best_params_)
```

Оформим результат оценки качества моделей в таблицу:

```python
score_data = pd.DataFrame(data=[10.1, 9.27, 8.72], 
                       index=['Линейная регрессия', 'Дерево решений', 'Случайный лес'],
                       columns=['Значение sMAPE'])

display(score_data.sort_values('Значение sMAPE'))
```

Лучший результат показала модель *RandomForestRegressor (random_state=STATE, n_jobs=-1, n_estimators=22 , max_depth=3)*. Значение *sMAPE* при данных гиперпараметрах равно 8.72

### Тестирование модели

Проведем проверку выбранной нами модели RandomForestRegressor *(random_state=STATE, n_jobs=-1, n_estimators=22, max_depth=3)* на тестовой выборке:

```python
model = RandomForestRegressor (random_state=STATE, n_jobs=-1, n_estimators=22 , max_depth=3)

model.fit(features_train, target_train)
predicted_test = model.predict(features_test) # предсказания модели на тестовой выборке
score = final_smape(target_test, predicted_test) # значение метрики sMAPE на тестовой выборке
print('Значение sMAPE на тестовой выборке равно', score)
```

Значение *sMAPE* нашей модели на тестовой выборке равно 7.33, что выше, чем на тренировочной (*sMAPE* = 8.72). Значит, наша модель не переобучилась и успешно прошла тест. 

### Проверка модели на адекватность

Проверим модель на адекватность с помощью DummyRegressor - случайного регрессора с настройкой на постоянное предсказание среднего значения обучающей выборки:

```python
dummy_model = DummyRegressor(strategy='mean')
dummy_model.fit(features_train, target_train)
dummy_predicted = dummy_model.predict(features_test)
score = final_smape(target_test, dummy_predicted)

print('Значение sMAPE константной модели на тестовой выборке равно', score)
```

Проверка на адекватность пройдена - качество предсказания нашей модели выше, чем у константной модели (*sMAPE* равно 7.33 и 7.77 соответственно)
